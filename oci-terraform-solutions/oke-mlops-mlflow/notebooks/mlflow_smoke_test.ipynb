```python
# Import only what we need for a minimal logging demo
import os
import random
from datetime import datetime
import mlflow
```

```python
# Default to the currently deployed MLflow URL
tracking_uri = os.getenv("MLFLOW_TRACKING_URI", "http://mlflow.mlflow.svc.cluster.local")
mlflow.set_tracking_uri(tracking_uri)
print("Tracking URI:", tracking_uri)
```

```python
experiment_name = "oke-smoke-test"
# Create the experiment if it does not already exist
mlflow.set_experiment(experiment_name)
# Start a run and log a couple of sample values
with mlflow.start_run(run_name=f"local-smoke-{datetime.utcnow().strftime('%Y%m%d-%H%M%S')}"):
    accuracy = round(random.uniform(0.85, 0.99), 4)
    loss = round(random.uniform(0.01, 0.2), 4)
    mlflow.log_param("source", "local-notebook")
    mlflow.log_param("model", "demo-classifier")
    mlflow.log_metric("accuracy", accuracy)
    mlflow.log_metric("loss", loss)
    print("Logged run successfully")
    print("accuracy:", accuracy)
    print("loss:", loss)
```

```python
# Quick check to confirm the experiment is visible in tracking server
client = mlflow.MlflowClient()
exp = client.get_experiment_by_name("oke-smoke-test")
print("Experiment ID:", exp.experiment_id if exp else "not found")
```
